```python
#converters=None
#Assuming 'Unit Cost' is in USD, converting it to INR (USDINR=72) while reading

exchange_rate = 74.5

def convert_to_inr(x):
    return x * exchange_rate

sampledata_df3 = pd.read_excel('/home/lubuntu/Downloads/dataset/SampleData.xlsx', sheet_name='SalesOrders', parse_dates=[1], convert_float=False, converters={'Unit Cost':convert_to_inr})

sampledata_df3.head()

```
